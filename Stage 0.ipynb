```python
from collections import Counter


def load_ciphertext(file_path):
    """Load the ciphertext from a specified file."""
    try:
        with open(file_path, 'r') as file:
            return file.read()
    except FileNotFoundError:
        print(f"Error: {file_path} not found.")
        return None


def frequency_distribution(ciphertext):
    """Calculate and display the frequency distribution of characters."""
    cleaned_text = ''.join(filter(str.isalpha, ciphertext)).upper()
    frequency_map = Counter(cleaned_text)
    print("Frequency Distribution:")
    for letter, count in frequency_map.most_common():
        print(f"{letter}: {count}")


def create_substitution_map(ciphertext, letters):
    """Create a mapping from cipher characters to plain characters based on frequency."""
    cleaned_text = ''.join(filter(str.isalpha, ciphertext)).upper()
    freq_analysis = Counter(cleaned_text)
    sorted_cipher_letters = [item[0] for item in freq_analysis.most_common()]
    return dict(zip(sorted_cipher_letters, letters[:len(sorted_cipher_letters)]))


def decrypt(ciphertext, cipher_to_plain):
    """Decrypt the ciphertext using the provided mapping."""
    return ''.join(cipher_to_plain.get(char, char) for char in ciphertext.upper())


def main():
    # Configuration
    file_path = 'stage_0.txt'
    letters = "ETAOINSHRLDUMYWCGFBPKVZJXQ"  # English letter frequency order

    # Load the ciphertext
    ciphertext = load_ciphertext(file_path)
    if ciphertext is None:
        return

    # Display frequency distribution
    frequency_distribution(ciphertext)

    # Create a substitution mapping
    cipher_to_plain = create_substitution_map(ciphertext, letters)

    # Decrypt the ciphertext
    decrypted_text = decrypt(ciphertext, cipher_to_plain)
    print("\nDecrypted Text:", decrypted_text)


if __name__ == "__main__":
    main()
```
